```python
data = [{ "Name": "Alice", "Age": 25, "City": "New York", "Salary" : 70000},
        { "Name": "Bob", "Age": 31, "City": "San Francisco", "Salary" : 80000},
        { "Name": "Charlie", "Age": 35, "City": "Los Angeles", "Salary" : 90000}]

# What if we want to analyze this data?

# Maybe look at average salary, or maybe find people older than 30?

total_sl =0
count_sl =0
for person in data:
    total_sl += person['Salary']
    count_sl +=1
    av_salary = total_sl/count_sl
print(f"av_salary is: {av_salary}")

older_than_30 = []
for person in data:
    if person["Age"]>30:
        older_than_30.append(person)
    

print(older_than_30)




```

```python
import pandas as pd
import numpy as np
# find people older than 30

daf = pd.DataFrame(data)
av_salary = daf['Salary'].mean()
older_than_30 = daf[daf['Age']>30]
print(av_salary)
print(older_than_30)
```
