```python
# What proble does pandas solve?

data = [{'name': 'Alis', 'age': 25, 'city': 'New York', 'Salary': 70000},
        {'name': 'Bob', 'age': 31, 'city': 'San Francisco', 'Salary': 80000},
        {'name': 'Charlie', 'age': 35, 'city': 'Los Angeless', 'Salary': 90000}]

# What if we want analys this data?
# Maybe look att average  salary or maybe findout people older than 30

total = 0
count = 0
for person in data:
    total += person['Salary']
    count += 1
average_salary = total / count

print('\n Average salary is', average_salary)

older_than_30 = []
for person in data:
    if person['age'] > 30:
        older_than_30.append(person)

print('\n People older than 30', older_than_30)
```

```python
import pandas as pd

df = pd.DataFrame(data)

print('\nDataFrame:')
print(df)

average_salary = df['salary'].mean()
older_than_30 = df[df['age']>30]

print('\n Average salary:', average_salary)
print('People older than 30:\n', older_than_30)
```

Data cleaning with pandas: Handling Missing Data, Duplicat and Data transformation
Paandas is very powerful for cleaning and data manupilation

```python
data_with_nan = {
    'name': ['Alis', 'Bob', 'Charlie', None, 'Eve'],
    'age': [25, None, 35, 28, 32],
    'city': ['New York', 'London', None, 'Paris', 'Sydney'],
    'salary': [50000, 60000, 70000, 55000, None]
}

df_nan = pd.DataFrame(data_with_nan)
print('\nDataframe with nan is:')
print(df_nan)


# Check for missing values

print('\nMissing values in each column:\n', df_nan.isnull().sum())


# Fill missing values
# we fill missing values with fillna()-method
df_filled = df_nan.fillna({
    'name': 'Unknow',
    'age': df_nan['age'].mean(),
    'city': 'Unknown',
    'salary': df_nan['salary'].mean()
})

print('\nDataFrame after filling missing values:\n', df_filled)

# Drop rows with missing values

df_dropped = df_nan.dropna()
print('\nDataFrame after dropping rows with missing values:\n', df_dropped)
```

```python
# isnull find all the NaN+None in the DataFrame
print(df_nan.isnull())

# Then we can use sum() to count the number True values in each column
print('\nCount of missing values in each column:\n', df_nan.isnull().sum())
```

Sorting/Filtering Data

```python
# Sorting by a single column

print('Sorted by age:\n', df_filled.sort_values(by='age') )

print('Sorted by salary Descending:\n', df_filled.sort_values(by='salary', ascending=False) )

# Multiple criteria sorting 
print('\nSorted by city, then by age:\n', df_filled.sort_values(by=['city', 'age']))

# Complex filtering using multiple boolean conditions
print('High earning adults in their 30s:\n', df_filled[(df_filled['age'] >=30) & (df_filled['salary'] >= 60000)])
```

Grouping Aggregating Data

```python
print(df_filled)

city_group = df_filled.groupby('city')

city_group['salary'].mean()

city_stats = df_filled.groupby('city').agg({
    'age': ['mean', 'min', 'max'],
    'salary': ['mean', 'sum', 'count']
})

print('\nCity-wise statistics:\n', city_stats)
```
